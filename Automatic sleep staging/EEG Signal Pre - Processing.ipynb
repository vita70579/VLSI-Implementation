* 20 個受試者 (Subjects),每個受試者有兩個夜晚的紀錄檔,其中 subject13只有一個夜晚。

```python
record_name = []
annot_name = []

events = []  # events
for file_name in os.listdir(recording_path):
    match1 = re.search(record_regu, file_name)
    match2 = re.search(annot_regu, file_name)
    if (match1):
        record_name.append(match1.group())
    elif (match2):
        annot_name.append(match2.group())
    
    del match1,match2,file_name
gc.collect()

if (len(record_name) != len(annot_name)):
    record_name = []
    annot_name = []

df = pd.concat([pd.DataFrame(record_name), pd.DataFrame(annot_name)], axis=1)
df.columns = ['Records', 'Hypnogram']
df
```

```python
raw_ = []  # raw
events_ = []  # events

for ren, an in zip(record_name, annot_name):
    print('\n\033[5;31;47m File name:\n %s\033[0m\n' % (ren))
    rec = mne.io.read_raw_edf(recording_path + ren, preload=False)
    ann = mne.read_annotations(recording_path + an)

    # set annotations
    rec.set_annotations(ann, emit_warning=True)
    rec.set_channel_types(channel_types)
    ev, _ = mne.events_from_annotations(rec,
                                        use_rounding=False,
                                        event_id=sleep_stage,
                                        chunk_duration=ep_duration)

    events_.append(ev)
    raw_.append(rec)
    
    del ren,an,ann,ev
gc.collect()
    
```

```python
epochs_ = []
tmax = ep_duration - 1/ rec.info['sfreq']
for raw, eve in zip(raw_, events_):
    ep = mne.Epochs(
        raw=raw,
        picks=['eeg'],
        events=eve,
        event_id=event_id,
        tmin=0.,
        tmax=tmax,
        baseline=None,  # 1 epoch data - mean ( 1 epoch data)
        detrend=None  # linear/constant detrend of 1 epoch
    )
    epochs_.append(ep)
    
    del raw,eve
del rec
gc.collect()
```

# PREPROCESSING

```python
def Filter(data,LH_freq,fs=100):
    sos = signal.butter(N=8,Wn=LH_freq,btype='bandpass',output='sos',fs=100)
    data = signal.sosfilt(sos, x=data)
    
    return data
```

```python
ep_df_ = []
EpIdx_ = []
EpIdx = pd.read_csv('EpIdx.csv', header=None)
i = 0
for eps in epochs_:
    df = eps.to_data_frame()

    # data balance
    if (ep_duration == 10):
        con_len = list(
            df.groupby(level=['condition']).count()['EEG Fpz-Cz'] /
            (ep_duration * sfreq))  # epochs
        drop_len = np.floor(con_len[-1] - np.mean(con_len[:4]))  # epochs

        star = ((int(np.floor(drop_len / 2) * (ep_duration * sfreq))) //
                int(int(smoothing_factor) * ep_duration * sfreq)) * int(
                    int(smoothing_factor) * ep_duration * sfreq)
        end = int(
            (np.sum(con_len) - np.floor(drop_len / 2)) * (ep_duration * sfreq))
        end = end - (
            (end - star) % int(smoothing_factor * ep_duration * sfreq))
        EpIdx_.append((star, end))
    else:
        star = EpIdx[0][i]
        end = EpIdx[1][i]
        i += 1
    df = df[star:end]

    # Basic Filter
    df['EEG Fpz-Cz'] = Filter(data=df['EEG Fpz-Cz'],
                              LH_freq=[0.5, 30],
                              fs=sfreq)
    df['EEG Fpz-Cz Delta'] = Filter(data=df['EEG Fpz-Cz'],
                                    LH_freq=[0.5, 4.5],
                                    fs=sfreq)
    df['EEG Fpz-Cz Theta'] = Filter(data=df['EEG Fpz-Cz'],
                                    LH_freq=[4.5, 8.5],
                                    fs=sfreq)
    df['EEG Fpz-Cz Alpha'] = Filter(data=df['EEG Fpz-Cz'],
                                    LH_freq=[8.5, 11.5],
                                    fs=sfreq)
    df['EEG Fpz-Cz Sigma'] = Filter(data=df['EEG Fpz-Cz'],
                                    LH_freq=[11.5, 15.5],
                                    fs=sfreq)
    df['EEG Fpz-Cz Beta'] = Filter(data=df['EEG Fpz-Cz'],
                                   LH_freq=[15.5, 30],
                                   fs=sfreq)

    ep_df_.append(df)
    del df, star, end
gc.collect()
if (ep_duration == 10):
    pd.DataFrame(EpIdx_).to_csv('EpIdx.csv', header=None, index=None)
```

```python
def Block_index(DataRecordInfo,dataframe_index,train_or_test = 'training'):
    start_idx_ = []
    idx = 0
    start_idx_.append(idx)
    
    for i in dataframe_index:
        idx = idx + (DataRecordInfo['total'][i]/smoothing_factor)
        start_idx_.append(int(idx))
    
    df = pd.DataFrame(start_idx_)
    df.columns = [train_or_test+'block start index']
    df.to_csv(train_or_test+'_block_start_index.csv',index=0)
    print("Total "+train_or_test+" Size (30s/epoch):",np.sum(DataRecordInfo['total'][dataframe_index])/smoothing_factor)
    print(df) # df中的索引是劃分完後,每個 Subject 開始的 Epoch 在劃分集中的索引。(30 s/Ep)
    
def Block_Epochs(DataRecordInfo,dataframe_index,train_or_test):
    df = pd.DataFrame((DataRecordInfo['total'][dataframe_index]/smoothing_factor)).astype(np.int).reset_index(drop=True)
    df.columns = [train_or_test+' records epochs']
    df.to_csv(train_or_test+'_records_epochs.csv',index=0)
    print(df) # df中的索引是劃分完後,每個 Subject 開始的 Epoch 在劃分集中的索引。(30 s/Ep)
```

```python
# multi processing
def applyParallel(dfGrouped, func):
    retLst = Parallel(n_jobs=mp.cpu_count())(delayed(func)(group)
                                             for name, group in dfGrouped)
    return pd.concat(retLst, axis=1).T
```

```python
def Features_Calculate(df):
    grouped = df.groupby(level=['epoch', 'condition'])
    temps = []
    midx = []

    # multiIndex
    for idx, gr in grouped:
        midx.append(idx)
    midx = pd.MultiIndex.from_tuples(midx, names=('epoch', 'condition'))
    """======================================================================================================================="""
    temps.append(grouped.std())
    temps.append(
        applyParallel(
            grouped, lambda x: pd.Series(univariate.compute_ptp_amp(x.T.values)
                                         )).set_index(midx))

    temps.append(
        applyParallel(
            grouped,
            lambda x: pd.Series(univariate.compute_zero_crossings(x.T.values)
                                )).set_index(midx))
    temps.append(
        applyParallel(
            grouped,
            lambda x: x.apply(Power, sfreq=100, axis=0)).set_index(midx))

    temps.append(
        applyParallel(grouped,
                      lambda x: x.apply(Spect_slope, axis=0)).set_index(midx))

    temps.append(
        applyParallel(
            grouped,
            lambda x: x.apply(Mean_frequency, axis=0)).set_index(midx))
    temps.append(
        applyParallel(
            grouped, lambda x: pd.Series(
                univariate.compute_hjorth_complexity(x.T.values))).set_index(
                    midx))
    temps.append(
        applyParallel(
            grouped,
            lambda x: x.apply(spectral_entropy,sf=100, method='welch')).set_index(
                midx))
    temps.append(
        applyParallel(grouped,
                      lambda x: x.apply(perm_entropy)).set_index(midx))
    temps.append(
        applyParallel(
            grouped, lambda x: x.apply(
                svd_entropy, order=3, delay=1, normalize=False, axis=0)).
        set_index(midx))
    temps.append(
        applyParallel(
            grouped,
            lambda x: x.apply(sample_entropy, order=2, axis=0)).set_index(
                midx))
    temps.append(
        applyParallel(
            grouped, lambda x: x.apply(AutoCorr, k=1, axis=0)).set_index(midx))

    """================================================================================"""

    res = pd.concat(temps, axis=1)
    cols = [(i, j) for i in funcs_n for j in six_bands_n]
    res.columns = pd.MultiIndex.from_tuples(tuples=cols,
                                            names=['Function', 'Band'])

    return res
```

```python
temp_1 = []

st = time.time()
for df in tqdm(ep_df_):
    temp_1.append(Features_Calculate(df))
    del df

gc.collect()
et = time.time()
```

```python
tr_df = pd.concat(temp_1, axis=0)
print('time-consuming: %f' % (et - st))
```

# FEATURE CROSSING

```python
# Feature cross
lev1_comblist = list(combinations(six_bands_n, r=2))

for fes in funcs_n:
    cInd = int((funcs_n.index(fes) + 1) * len(six_bands_n) +
               funcs_n.index(fes) * len(ratio_n))

    i = 0
    for rn, coms in zip(ratio_n, lev1_comblist):
        ncol = (fes, rn)
        nval = tr_df[(fes, coms[1])] / tr_df[(fes, coms[0])]
        tr_df.insert(loc=(cInd + i), column=ncol, value=nval)
        i = i + 1

        del rn, coms, ncol, nval
    del fes
gc.collect()
tr_df.head()
```
